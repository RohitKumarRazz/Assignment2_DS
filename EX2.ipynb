```python
import matplotlib.pyplot as plt

x=[2,4,6,8]
y=[80,60,40,20]

plt.plot(x, y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Line plot')
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Scatter plot')
plt.show()
```

```python
categories = ['Azharuddin','Md juned','Parwej','MD Ajbullah','Vivek','Rana']
values = [90, 40, 20, 65 , 50, 55]
plt.bar(categories, values)
plt.xlabel('Names')
plt.ylabel('Capacity')
plt.title('Bar plot')
plt.show()
```

```python
data = [2,4,6,8,20,40,60,40,70,80]
plt.hist(data, bins=6)
plt.xlabel('values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()

```

```python
labels = ['Label 1', 'Label 2', 'Label 3', 'Label 4','Label 5','Label 6']
sizes = [30, 20, 15, 5, 10, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('pie chart')
plt.show()

```
